```python
salt_dome = gp.create_model('salt_dome_s33')
gp.init_data(salt_dome, [-2000, 6000., -2000, 6000., 0, 5000.], [50, 50, 50],
             path_o= data_path + "\salt_dome_s33_bottom\salt_dome_s33_bottom_O.csv",
             path_i= data_path + "\salt_dome_s33_bottom\salt_dome_s33_bottom_I.csv",
             default_values=True)
```

```python
from gempy.plot.vista import GemPyToVista
viewup = [0, 0, 1]
gpv = GemPyToVista(salt_dome)
gpv.plot_surface_points()
gpv.plot_orientations()
gpv.plot_surfaces()
p = gpv.p
p.camera_position = 'xz'
p.camera.elevation = 45
p.add_bounding_box()
p.set_scale(zscale=1)
path = p.generate_orbital_path(n_points=36,factor=2.0,viewup=viewup, shift=0.2)
p.open_gif("s33_bottom.gif")
p.orbit_on_path(path, write_frames=True,viewup=viewup, step=0.05)
p.close()
#gpv.p.notebook='True'
gpv.p.show()
```
