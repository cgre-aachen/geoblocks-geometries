```python
import gempy as gp
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import pyvista as pv
data_path = 'D:\sciebo\geoblk\work-flow\geoblocks-workflow\examples\example_data'
```

```python
salt_dome = gp.create_model('salt_dome_s32')
gp.init_data(salt_dome, [-4000, 10000., -7000, 10000., 0, 5000.], [50, 50, 50],
             path_o= data_path + "\salt_dome_s46_right\salt_dome_s46_right_O.csv",
             path_i= data_path + "\salt_dome_s46_right\salt_dome_s46_right_I.csv",
             default_values=True)
```
